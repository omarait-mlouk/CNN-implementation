```python
# CNN Implementation using Keras - MNIST Classification
# Facult polydisciplinaire de Ouarzazate - Master IAA & IMSD

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.utils import to_categorical
from keras.optimizers import Adam
import seaborn as sns

# =============================================================================
# Step 1: Load Data and Visualization
# =============================================================================

print("Step 1: Loading MNIST dataset and visualization")
print("=" * 50)

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"Training data shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# Display class distribution using pandas
train_df = pd.DataFrame({'labels': y_train})
class_distribution = train_df['labels'].value_counts().sort_index()
print(f"\nClass distribution in training set:")
print(class_distribution)

# Visualize class distribution
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
class_distribution.plot(kind='bar')
plt.title('Class Distribution in Training Set')
plt.xlabel('Digit Class')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)

# Display sample images
plt.subplot(1, 2, 2)
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
fig.suptitle('Sample MNIST Images')

for i in range(10):
    row = i // 5
    col = i % 5
    # Find first occurrence of each digit
    idx = np.where(y_train == i)[0][0]
    axes[row, col].imshow(X_train[idx], cmap='gray')
    axes[row, col].set_title(f'Digit: {i}')
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()

# =============================================================================
# Step 2: Model Designing
# =============================================================================

print("\nStep 2: Model designing")
print("=" * 50)

# Set hyperparameters
BATCH_SIZE = 128
EPOCHS = 10
NUM_CLASSES = 10
INPUT_SHAPE = (28, 28, 1)

# Preprocess the data
# Normalize pixel values to [0, 1]
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Reshape data to add channel dimension
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

# Convert labels to categorical
y_train_cat = to_categorical(y_train, NUM_CLASSES)
y_test_cat = to_categorical(y_test, NUM_CLASSES)

print(f"Preprocessed training data shape: {X_train.shape}")
print(f"Preprocessed test data shape: {X_test.shape}")

# Build CNN model according to specifications
# Note: Adding padding to prevent dimension issues
model = Sequential([
    # conv1: 32 convolutions 55, followed by ReLU
    Conv2D(32, (5, 5), activation='relu', padding='same', input_shape=INPUT_SHAPE, name='conv1'),

    # pool1: max-pooling 22
    MaxPooling2D((2, 2), name='pool1'),

    # conv2: 64 convolutions 55, followed by ReLU
    Conv2D(64, (5, 5), activation='relu', padding='same', name='conv2'),

    # pool2: max-pooling 22
    MaxPooling2D((2, 2), name='pool2'),

    # conv3: 64 convolutions 55, followed by ReLU
    Conv2D(64, (5, 5), activation='relu', padding='same', name='conv3'),

    # pool3: max-pooling 22
    MaxPooling2D((2, 2), name='pool3'),

    # Flatten for fully connected layers
    Flatten(),

    # fc4: fully-connected, 100 neurons output, followed by ReLU
    Dense(100, activation='relu', name='fc4'),
    Dropout(0.5),  # Add dropout for regularization

    # fc5: fully-connected, 10 neurons output, followed by softmax
    Dense(NUM_CLASSES, activation='softmax', name='fc5')
])

# Alternative approach: Use smaller kernel sizes for later layers
# Uncomment this if you want to stick closer to the original specification without padding
"""
model = Sequential([
    # conv1: 32 convolutions 55, followed by ReLU
    Conv2D(32, (5, 5), activation='relu', input_shape=INPUT_SHAPE, name='conv1'),

    # pool1: max-pooling 22
    MaxPooling2D((2, 2), name='pool1'),

    # conv2: 64 convolutions 55, followed by ReLU
    Conv2D(64, (5, 5), activation='relu', name='conv2'),

    # pool2: max-pooling 22
    MaxPooling2D((2, 2), name='pool2'),

    # conv3: 64 convolutions 33, followed by ReLU (reduced kernel size)
    Conv2D(64, (3, 3), activation='relu', name='conv3'),

    # pool3: max-pooling 22
    MaxPooling2D((2, 2), name='pool3'),

    # Flatten for fully connected layers
    Flatten(),

    # fc4: fully-connected, 100 neurons output, followed by ReLU
    Dense(100, activation='relu', name='fc4'),
    Dropout(0.5),

    # fc5: fully-connected, 10 neurons output, followed by softmax
    Dense(NUM_CLASSES, activation='softmax', name='fc5')
])
"""

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Display model architecture
print("\nModel Architecture:")
model.summary()

# =============================================================================
# Step 3: Train
# =============================================================================

print("\nStep 3: Training the model")
print("=" * 50)

# Train the model
history = model.fit(
    X_train, y_train_cat,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_test_cat),
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# =============================================================================
# Step 4: Test
# =============================================================================

print("\nStep 4: Testing the model")
print("=" * 50)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Make predictions
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1)

# Display some test results
plt.figure(figsize=(12, 8))
for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(X_test[i].reshape(28, 28), cmap='gray')
    plt.title(f'True: {y_test[i]}, Pred: {predicted_classes[i]}')
    plt.axis('off')

plt.suptitle('Test Results - True vs Predicted Labels')
plt.tight_layout()
plt.show()

# Confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

cm = confusion_matrix(y_test, predicted_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, predicted_classes))

print("\nKeras CNN Implementation Complete!")
print("=" * 50)
```
